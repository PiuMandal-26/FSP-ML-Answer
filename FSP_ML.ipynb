```python
#Q.1
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler

data = {'Marks': [80, 90, np.nan, 70, np.nan]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

print("\nMissing Values:")
print(df.isnull())

df_mean = df.fillna(df['Marks'].mean())
print("\nAfter Replacing With Mean:")
print(df_mean)

df_median = df.fillna(df['Marks'].median())
print("\nAfter Replacing With Mean:")
print(df_median)

df_drop = df.dropna()
print("\nAfter Dropping Missing Rows:")
print(df_drop)
```

```python
#Q.2
import pandas as pd

data = {
    'ID': [1, 2, 3, 2, 4, 3],
    'Name': ['Alice', 'Bob', 'Charlie', 'Bob', 'David', 'Charlie'],
    'Marks': [85, 90, 78, 90, 88, 78]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

duplicates = df.duplicated()

print("\nDuplicate Entries:")
print(duplicates)

print("\nRows that are Duplicates:")
print(df[df.duplicated()])

df_no_duplicates = df.drop_duplicates()

print("\nDataset after Removing Duplicate Rows:")
print(df_no_duplicates)
```

```python
#Q.3
values = [12, 25, 40, 55, 70]

x_min = min(values)
x_max = max(values)

normalized_values = [(x - x_min) / (x_max - x_min) for x in values]

print("Original Values:", values)
print("Normalized Values:", normalized_values)
```

```python
#Q.4
import numpy as np

data = [12, 25, 40, 55, 70]

mean = np.mean(data)
std_dev = np.std(data)

z_scores = [(x - mean) / std_dev for x in data]

print("Original Data:", data)
print("Mean =", mean)
print("Standard Deviation =", std_dev)
print("Z-score Normalized Data:")

for x, z in zip(data, z_scores):
    print(f"{x} -> {z:.4f}")
```

```python
#Q.5
import numpy as np

data = [10, 12, 14, 15, 18, 20, 22, 100]

mean = np.mean(data)
std_dev = np.std(data)

outliers = []

print("Data\tZ-Score")
for x in data:
    z_score = (x - mean) / std_dev
    print(f"{x}\t{z_score:.2f}")

    if abs(z_score) > 2:   # Threshold can be 2 or 3
        outliers.append(x)

print("\nMean =", mean)
print("Standard Deviation =", std_dev)
print("Outliers:", outliers)
```

```python
#Q.6
from sklearn.preprocessing import LabelEncoder

colors = ["Red", "Blue", "Green", "Blue"]

print("Original Data:")
print(colors)

label_encoder = LabelEncoder()

encoded_colors = label_encoder.fit_transform(colors)

print("\nEncoded Data:")
print(encoded_colors)

print("\nLabel Mapping:")
for color, label in zip(label_encoder.classes_,
                        range(len(label_encoder.classes_))):
    print(f"{color} -> {label}")
```

```python
#Q.7
import pandas as pd

data = {'City': ['Kolkata', 'Delhi', 'Mumbai', 'Chennai']}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

encoded_df = pd.get_dummies(df, columns=['City'])

print("\nOne-Hot Encoded Dataset:")
print(encoded_df)
```

```python
#Q.8
import pandas as pd
from sklearn.preprocessing import StandardScaler

df= pd.DataFrame({
    'Height': [150, 160, 170, 180],
    'Weight': [50,60,70,80],
    'Age': [20,30,40,50]
})

scaler = StandardScaler()
df_scaled = pd.DataFrame(
    scaler.fit_transform(df),
    columns=df.columns
)
print(df_scaled.round(4))
```

```python
#Q.9
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

data = {
    'Values': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

minmax_scaler = MinMaxScaler()
df['MinMax_Scaled'] = minmax_scaler.fit_transform(df[['Values']])

standard_scaler = StandardScaler()
df['Standardized'] = standard_scaler.fit_transform(df[['Values']])

print("\nDataset after Scaling:")
print(df)
```

```python
#Q.10
import pandas as pd

data = {
    'Numbers': ['10', '20', '30', '40', '50']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\nData Type Before Conversion:")
print(df.dtypes)

df['Numbers'] = df['Numbers'].astype(int)

print("\nDataFrame After Conversion:")
print(df)
print("\nData Type After Conversion:")
print(df.dtypes)
```
